```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [1, 3, 2]

f = interp1d(x, y)

y_hat = f(1.5)
print(y_hat)
```

```python
plt.figure(figsize=(10, 8)) #ukuran plot 10x8 inci
plt.plot(x, y, '-ob') # menampilkan titik data dgn garis biru dan lingkaran
plt.plot(1.5, y_hat, 'ro') #menampilkan titik hasil interpolasi dengan titik merah
plt.title('Linear interpolation at x = 1.5') # judul grafik
plt.xlabel('x') # label sumbu x
plt.ylabel('y') # label sumbu y
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def divided_diff(x, y):
  n = len(y)
  coef = np.zeros([n, n]) #membuat tabel kosong untuk divided differences
  coef[:, 0] = y #kolom pertama berisi nilai y

  for j in range(1, n): #loop untuk mengisi kolom-kolom berikutnya
    for i in range(n - j):
      coef[i][j] = (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j] - x[i])

  return coef
def newton_poly(coef, x_data, x):
  n = len(x_data) - 1
  p = coef[n]

  for k in range(1, n+1):
    p = coef[n-k] + (x - x_data[n-k])*p

  return p
```

```python
x = np.array([-5, -1, 0, 2])
y = np.array([-2, 6, 1, 3])

#menghitung keofisien divided difference
a_s = divided_diff(x, y)[0, :]

#evaluasi interpolasi pada titik-titik baru
x_new = np.arange(-5, 2.1, .1)
y_new = newton_poly(a_s, x, x_new)

#membuat plot
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo') #menampilkan titik data asli
plt.plot(x_new, y_new) #menampilkan kurva interpolasi
```

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2] #titik data x
y = [1, 3, 2] #titik data y

#membuat interpolasi kubik dengan batas natural
f = CubicSpline(x, y, bc_type='natural')

#membuat rentaing nilai x untuk plot hasil interpolasi
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)

#membuat plot
plt.figure(figsize=(10, 8)) #ukuran plot
plt.plot(x_new, y_new, 'b') #menampilkan kurva interpolasi(biru)
plt.plot(x, y, 'ro') #menampilkan titik data asli(merah)
plt.title('Cubic Spline Interpolation') #judul
plt.xlabel('x') #label x
plt.ylabel('y') #labe y
plt.show()
```
